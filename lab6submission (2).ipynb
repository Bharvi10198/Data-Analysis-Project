```python
#Quesition 1
movies=pd.read_csv('movies.csv')
movies.head()
```

```python
financials=pd.read_csv('financials.csv')
financials.head()
```

```python
movies_financials = movies.merge(financials, on='id', how='left')
missing_count = movies_financials['budget'].isna().sum()
print(f"Total movies missing financial details: {missing_count}")
```

```python
# Question 2
taglines=pd.read_csv('taglines.csv')
```

```python
left_merged_df = movies_financials.merge(taglines, on="id", how="left")
inner_merged_df = movies_financials.merge(taglines, on="id", how="inner")
left_join_count = len(left_merged_df)
inner_join_count = len(inner_merged_df)
print(f"Number of rows after the left join: {left_join_count}")
print(f"Number of rows after the inner join: {inner_join_count}")
```

```python
# Question 3
genre=pd.read_csv('movie_to_genres.csv')
genre_counts_df = genre.groupby("movie_id")["genre"].nunique()
sci_fi_movies_df = genre[genre["genre"] == "Science Fiction"]
sci_fi_movies_df = sci_fi_movies_df[sci_fi_movies_df["movie_id"].isin(genre_counts_df[genre_counts_df == 1].index)]
sci_fi_movies_df = sci_fi_movies_df.merge(movies, left_on="movie_id", right_on="id", how="left")
sci_fi_movies_df = sci_fi_movies_df[["id", "title", "release_date", "popularity"]]
print(sci_fi_movies_df.head())
```

```python
# Question 4
iron1=pd.read_csv('iron_1_actors.csv')
iron2=pd.read_csv('iron_2_actors.csv')
merged_actors_df = iron1.merge(iron2, on="id", how="outer", suffixes=("_iron1", "_iron2"))
unique_actors_df = merged_actors_df[merged_actors_df.isna().any(axis=1)]
unique_actors_df = unique_actors_df[["id", "name_iron1", "name_iron2"]]
print(unique_actors_df.head())
```
